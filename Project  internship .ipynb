```python
def Outlier_treatment(employee,col):
    Q1=employee[col].quantile(0.25)
    Q3=employee[col].quantile(0.75)
    IQR=Q3-Q1
    UW=Q3+1.5* IQR
    LW=Q1+1.5* IQR
    employee.loc[employee[col]>UW,col]=employee[col].median()
    employee.loc[employee[col]>LW,col]=employee[col].median()
    return employee
```

```python
for col in employee.select_dtypes(['int','float']):
    Outlier_treatment(employee,col)
```
